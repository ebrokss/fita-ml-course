# 1. nedēļa: ML problēmas definēšana

Šajā notebook izmantoju Titanic datu kopu, lai saprastu, ko tieši šeit vajadzētu prognozēt un kādi dati tam ir pieejami.

```python
import seaborn as sns

df = sns.load_dataset('titanic')

print(f"Rindu skaits: {df.shape[0]}")
print(f"Kolonnu skaits: {df.shape[1]}")
df.head()
```

## A. Problēmas apraksts

Šajā datu kopā gribu prognozēt, kādi dati norāda uz to, kas pasažieriem palīdzēja izdzīvot vai nē. To var mēģināt noteikt pēc tādām pazīmēm kā vecums, dzimums, biļetes klase un citiem datiem par pasažieriem. Manuprāt, šī problēma ir svarīga, jo tā palīdz saprast, kuri faktori visvairāk bija tieši saistīti ar izdzīvošanu. Šāds rezultāts būtu noderīgs analītiķim, kurš grib izvērtēt drošības riskus un pieņemt lēmumus, kuri ir balstīti uz datiem.


## B. ML uzdevuma tips

Es šo uzdevumu klasificēju kā **supervised learning**, jo datu kopā jau ir dota pareizā atbilde - kolonna `survived`. Tā ir **klasifikācijas** problēma, jo jāparedz viena no divām klasēm: `0` vai `1`, nevis kāda nepārtraukta skaitliska vērtība.


## C. Target variable

Target mainīgais šeit ir `survived`. Zemāk parādu, cik pasažieru izdzīvoja un cik neizdzīvoja, kā arī kopējo izdzīvojušo procentu.


```python
print(df['survived'].value_counts())
print(f'\nIzdzīvoja: {df["survived"].mean():.1%}')
```

Kolonna `survived` ir tā vērtība, kuru modelis mēģinātu paredzēt. `0` nozīmē, ka pasažieris neizdzīvoja, bet `1` nozīmē, ka pasažieris izdzīvoja.


## D. Features saraksts

Zemāk ir redzams features pārskats, sadalot tās skaitliskajās un kategoriskajās pazīmēs.


```python
print('Skaitliskās features:')
print([col for col in df.select_dtypes(include='number').columns.tolist() if col != 'survived'])
print('\nKategoriskās features:')
print(df.select_dtypes(exclude='number').columns.tolist())
```

## E. Sākotnējie novērojumi

Redzams, ka datu kopā ir gan skaitliskās, gan kategoriskās kolonnas, tāpēc vēlāk modelēšanā būtu jāprot strādāt ar dažādiem datu tipiem. Izdzīvoja apmēram 38.4% pasažieru, tātad lielākā daļa pasažieru neizdzīvoja. Vēl pamanīju, ka kolonna `alive` būtībā atkārto `survived` tekstuālā formā, tāpēc to nevajadzētu izmantot kā ievades pazīmi.
